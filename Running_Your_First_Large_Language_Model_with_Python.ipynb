# **Step 2: Load a Pretrained Language Model**

Now we will load a simple text-generation model.

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

result = generator("Artificial Intelligence will", max_length=30)

print(result)
```
